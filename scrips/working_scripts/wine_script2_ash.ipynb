```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
from pprint import pprint 
import sys
import os
import csv
```

```python
wine_csv = os.path.join("winedata_2016.csv")
```

```python
Country = []
Exports = []
Value = []
length = []
```

```python
# with open(udemy_csv, encoding='utf-8') as csvfile:
with open(wine_csv) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    for row in csvreader:
        # Add title
        Country.append(row[1])

        # Add price
        Exports.append(row[2])

        # Add number of subscribers
        Value.append(row[2])


        # Get length of the course to just a number
        new_length = row[6].split(" ")
        length.append(str(new_length[0]))

# Zip lists together
cleaned_csv = zip(Country, Exports, Value)

# Set variable for output file
output_file = os.path.join("web_final.csv")

#  Open the output file
with open(output_file, "w") as datafile:
    writer = csv.writer(datafile)

    # Write the header row
    writer.writerow(["Country", "Exports", "Value"])

    # Write in zipped rows
    writer.writerows(cleaned_csv)
```
